<h1> <b> <u> List Functions </u> </b> </h1>

```python
list1=[1,2,3,5,6,7,8]
print("Original List: ")
print(list1)
print("\nAppend 23: ")
list1.append(23)
print(list1)
print("\nInsert 34 at 6th index: ")
list1.insert(6,34)
print(list1)
print("\nExtend: [1,2,3]")
list1.extend([1,2,3])
print(list1)
print("\nPop: ")
list1.pop()
print(list1)
print("\nRemove 5: ")
list1.remove(5)
print(list1)
print("\nreverse fn: ")
list1.reverse()
print(list1)
print("\nCopying into l2: ")
l2=list1.copy()
print("l2: ",l2)
print("\nCount occurence of 3: ")
print(list1.count(3))
print("\nSort in Ascending Order: ")
list1.sort()
print(list1)
print("\nSort in Descending Order: ")
list1.sort(reverse=True)
print(list1)
print("\nAccessing Elements:")
for i in list1:
  print(i,end= " - ")
print("\nRemoving all items: ")
list1.clear()
print(list1)
```

```python
#Print Incompleted Tasks acc to priority and time req
a=[{"Name":"Bundle keeping","Priority":1,"Time req":2,"Completed":"False"},
   {"Name":"Grocery Shopping","Priority":1,"Time req":2,"Completed":"True"},
{"Name":"Washing Clothes","Priority":2,"Time req":1,"Completed":"False"},
   {"Name":"Technical work","Priority":2,"Time req":3,"Completed":"False"}]
s=[]
m=[]
def add(a):
  for i in a:
    if i["Completed"]=="False":
      m.append(i)
      s.append(str(i["Priority"])+str(i["Time req"]))
add(a)
l=[]
s.sort()
for i in m:
  for j in s:
    if j not in l:
      if int(j[0])==i["Priority"] and int(j[1])==i["Time req"]:
        print(i["Name"])
        l.append(j)


```

```python
#Python program to check whether the string is Symmetrical or Palindrome
a=str(input("Enter: "))
b=len(a)
b=b//2
if a[:b]==a[b:]:
  print("symmetric string")
if a[::]==a[::-1]:
  print("str is pallindrome")
```

```python

#Python program to check if a string has at least one letter and one number
a="hell1o"
m=0
c=0
for i in a:
  if i.isdigit():
    m=m+1
  elif i.isalpha():
    c=c+1
if m>0 and c>0:
  print("string contains both digit and alpha")
if m>0 and c==0:
  print("only digit")
if m==0 and c>0:
  print("only alpha")
```

```python
#Python | Count the Number of matching characters in a pair of string
a="abcdefjh"
b="bhdfjmcbx"
k=[]
l=[]
for i in a:
  k.append(i)
for i in b:
  k.append(i)
for s in k:
  if s not in l:
    print(s,k.count(s))
    l.append(s)
```

```python

#Python – Odd Frequency Characters
a="hehl.kjnzsfjhsbc"
k=[]
for i in a:
  if i not in k:
    k.append(i)
for i in k:
  if a.count(i)%2!=0:
    print(i)
```
